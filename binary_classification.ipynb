```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
df_train_data = pd.read_csv("./input_data/train.csv")
df_test_data = pd.read_csv("./input_data/test.csv")
df_train_data.isna().sum()
df_train_data['workclass'].fillna('filled_unemployed',inplace=True)
df_train_data['occupation'].fillna('filled_no_occupation',inplace=True)
df_train_data['native-country'].fillna('filled_no_country',inplace=True)
df_train_data
```

```python
work_class = df_train_data['workclass'].unique()
education = df_train_data['education'].unique()
marital_status = df_train_data['marital-status'].unique()
occupation = df_train_data['occupation'].unique()
relationship = df_train_data['relationship'].unique()
race = df_train_data['race'].unique()
gender = df_train_data['gender'].unique()
native_country = df_train_data['native-country'].unique()
col_list = [work_class,education,marital_status,occupation,relationship,race,gender,native_country]
list_ = ["workclass","education","marital-status","occupation","relationship","race","gender","native-country"]

for idx, col in enumerate(col_list):
    name = list_[idx]
    act_name = list_[idx]
    name=dict()
    for key,value in enumerate(col,start=1):
        name[value]=key
    df_train_data[act_name+'_new']=df_train_data[act_name].map(name)
    df_test_data[act_name+'_new']=df_test_data[act_name].map(name)

df_new_train_data = df_train_data[['age','workclass_new','education_new','educational-num','marital-status_new','occupation_new','relationship_new','race_new','gender_new','capital-gain','capital-loss','hours-per-week','native-country_new','income_>50K']]
df_new_test_data = df_test_data[['age','workclass_new','education_new','educational-num','marital-status_new','occupation_new','relationship_new','race_new','gender_new','capital-gain','capital-loss','hours-per-week','native-country_new']]
y = df_new_train_data[['income_>50K']]
x = df_new_train_data[df_new_train_data.columns[~df_new_train_data.columns.isin(['income_>50K'])]]
```

```python
lr = LogisticRegression(random_state = 42)  
lr.fit(x,y)
```

```python

pred = lr.predict(df_new_test_data)
pred
```

```python
out = pd.DataFrame(pred)
out.columns=['outcome']
out.index.name='id'
out
out.to_csv("C:/Users/TurningMachineBinaryClassification/input_data/out.csv")
```
